````python
data['Installs'].str[-1].value_counts()
````

````python
# regex=True : the letter("+")'s not Regular Expression
data['Installs'] = data['Installs'].str.replace('+', '', regex=True)
````

````python
data['Installs'] = data['Installs'].str.replace(',', '')
````

````python
data['Installs'] = data['Installs'].astype(int)
data.info()
````

````python
data['Price'].str[0].value_counts()
````

````python
data['Price'] = data['Price'].str.replace('$', '', regex=True)
````

````python
data['Price'] = data['Price'].astype(float)
data.info()
````

````python
data['Type'].value_counts()
````

````python
data.loc[data['Price']==0.0, 'Type'] = 'Free'
data.loc[data['Price']!=0.0, 'Type'] = 'Paid'
data['Type'].value_counts()
````

````python
data['Content Rating'].value_counts()
````

````python
data[data['Genres'].isnull()]
````

````python
# Found the app info in Playstore
data.loc[data['Category'].isnull(), 'Category'] = 'LIFESTYLE'
data.loc[data['Genres'].isnull(), 'Genres'] = 'Lifestyle'
data.info()
````

````python
data['Current Ver'].unique()
````

````python
data['Current Ver'].fillna('1.0.0', inplace=True)
data.info()
````
