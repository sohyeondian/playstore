```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('data/playstore_00.csv')
```

```python
data.head(10)
```

### TOP 10 of Reviews

```python
top_reviews = data.sort_values(by='Reviews', ascending=False)[:10]
top_reviews
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
sns.set(style="darkgrid", font_scale = 1.5)
sns.barplot(top_reviews['Reviews'], top_reviews['App'], alpha=0.8)
plt.title('Top10 of Reviews')
plt.show()
```

### About Categories

```python
group_cg = data.groupby('Category')
group_cg.mean().head()
```

```python
# total installs by categories
total = group_cg.sum()['Installs'].sort_values(ascending=False)[:10]
explode = [0] * 10
explode[0] = 0.1
plt.pie(x=total.values, labels=total.index, autopct='%.1f%%', shadow=True, explode=explode)
plt.title('Top 10 Most Installed Categories')
plt.show()
```
